```python
movies = movies[['id','title','genres','overview','keywords','cast','crew', 'vote_average', 'vote_count']]
movies
```

```python
movies = movies.dropna()
```

```python
movies.duplicated().sum()
```

```python
movies.isnull().sum()
```

```python
mean = movies['vote_average'].mean()
mean
```

```python
quan = movies['vote_count'].quantile(0.95)
quan
```

```python
Best_movies = movies[movies['vote_count'] >=quan ]
Best_movies.head(5)
```

```python
def weighted_rating(x):
  v = x['vote_count']
  r = x['vote_average']
  return (v/(v+quan) * r) + (quan/(quan+v) * mean)
```

```python
Best_movies['rating'] = Best_movies.apply(weighted_rating, axis=1)
Best_movies.head(5)
```

```python
Best_movies = Best_movies.sort_values('rating', ascending=False)
Best_movies = Best_movies[['id','title', 'rating']]
```

```python
Best_movies
```

```python
movies
```

```python
import ast
```

```python
movies.head(5)
```

```python
def convert(obj):
  l = []
  for i in ast.literal_eval(obj):
    l.append(i['name'])
  return l
```

```python
movies['genres'] = movies['genres'].apply(convert)
movies['keywords'] = movies['keywords'].apply(convert)
movies['cast'] = movies['cast'].apply(convert)
```

```python
movies.head(5)
```

```python
def fetch_director(obj):
    l = []

    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            l.append(i['name'])
            break
    return l
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies['crew']
```

```python
movies.sample(5)
```

```python
movies = movies.drop(columns=['vote_average', 'vote_count'])
```

```python
movies
```

```python
genres_movie_list = movies[['title','genres']]
genres_movie_list = genres_movie_list.merge(Best_movies, on="title")
genres_movie_list
```

```python
def remove_spacing(obj):
  l = []
  for i in obj:
    l.append(i.replace(" ",""))
  return l

```

```python
movies['genres'] = movies['genres'].apply(remove_spacing)
movies['keywords'] = movies['keywords'].apply(remove_spacing)
movies['cast'] = movies['cast'].apply(remove_spacing)
movies['crew'] = movies['crew'].apply(remove_spacing)
```

```python
movies.rename(columns = {'crew':'director'}, inplace = True)
```

```python
movies.sample(5)
```

```python
# combining all relevent data

movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['director']
```

```python
data = movies.drop(columns=['overview', 'genres', 'keywords', 'cast', 'director'])
```

```python
data
```

```python
data['tags'] = data['tags'].apply(lambda x:" ".join(x))
data.head(5)
```

```python
data['tags'] = data['tags'].apply(lambda x:x.lower())
```

**Feature selection part**

```python
# CountVectorizer
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000, stop_words='english')
```

```python
vector = cv.fit_transform(data['tags']).toarray()
```

```python
vector.shape
```

```python
# cosine_similarity

from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity
```

```python
data[data['title'] == "Inception"].index[0]
```

```python
distance = sorted(list(enumerate(similarity[96])), reverse=True, key = lambda x:x[1])
distance
```

```python
def recommend(movie):
  index = data[data['title'] == movie].index[0]
  distance = sorted(list(enumerate(similarity[index])), reverse=True, key = lambda x:x[1])

  for i in distance[1:6]:
    print(data.iloc[i[0]].title)
```

```python
recommend('Avatar')
```
